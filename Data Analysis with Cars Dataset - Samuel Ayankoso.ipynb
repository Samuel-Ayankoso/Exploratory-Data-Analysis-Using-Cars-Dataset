```python
# Drop irrelevant coloumns
car=car.drop(["Wheelbase","Length"],axis=1)
car.head()
```

```python
# renaming the column names
car=car.rename(columns={"EngineSize":"EngineSize","Horsepower":"Hp"})
car.head()
```

```python
# Total number of rows and columns
car.shape
```

```python
# Rows containing duplicate data
duplicate_rows_car=car[car.duplicated()]
print("Number of duplicate rows:",duplicate_rows_car.shape)
```

```python
car.count
```

```python
# drop the duplicates
car=car.drop_duplicates()
car.head()
```

```python
# Find the null values
print(car.isnull().sum())
```

```python
# drop the missing values
car=car.dropna()
car.head()
```

```python
print(car.isnull().sum())
```

```python
#Plotting a Histogram for number of cars per brand
car.Make.value_counts().nlargest(40).plot(kind="bar",figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel("Number of cars")
plt.xlabel("Make")
```

```python
# Find the relations between the variables
plt.figure(figsize=(20,10))
c=car.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```
