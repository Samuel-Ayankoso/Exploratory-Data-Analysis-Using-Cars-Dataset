# Exploratory Data Analysis (EDA) with Cars Dataset
* In this simple project, some basic functions in pandas and numpy were used for the exploration and analysis of the dataset.
* In addition, matplotlib and seaborn were used for visualization.
* The dataset can be downloaded here: https://github.com/sassoftware/sas-viya-programming/blob/master/data/cars.csv

```python
# Import packages
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
```

```python
# Read the car dataset
car = pd.read_csv("C:/Users/Samual/OneDrive - aucegypt.edu/Documents/DATA ANALYTICS/Cars Project/cars.csv")

# Print the first 5 rows of the dataset
car.head(5)
```

```python
# Print the last 5 rows of the dataset
car.tail(5)
```

```python
# Data type of each features in the dataset
car.dtypes
```

```python
# Finding the count, mean, std, min, max of each data column
car.describe()
```

```python
# Histogram plot
car.hist(figsize=(20,30))
```

```python
# Relationship between categorical and continous variable
sns.boxplot(x="EngineSize",y="Horsepower", data=car)
```

```python
# Plotting relatios between variables
sns.pairplot(car)
```

```python
# Drop irrelevant coloumns
car=car.drop(["Wheelbase","Length"],axis=1)
car.head()
```

```python
# renaming the column names
car=car.rename(columns={"EngineSize":"EngineSize","Horsepower":"Hp"})
car.head()
```

```python
# Total number of rows and columns in the car dataset
car.shape
```

```python
# Rows containing duplicate data
duplicate_rows_car=car[car.duplicated()]
print("Number of duplicate rows:",duplicate_rows_car.shape)
```

```python
# drop the duplicates
car=car.drop_duplicates()
car.head()
```

```python
# Print the number of null values in each coulmn
print(car.isnull().sum())
```

```python
# drop the missing values
car=car.dropna()
car.head()
```

```python
# Print the number of null values in each coulmn
print(car.isnull().sum())
```

```python
#Plotting a Histogram for number of cars per brand
car.Make.value_counts().nlargest(40).plot(kind="bar",figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel("Number of cars")
plt.xlabel("Make")
```

```python
# The relations between the variables
plt.figure(figsize=(20,10))
c=car.corr()
sns.heatmap(c,cmap="BrBG",annot=True)
c
```
